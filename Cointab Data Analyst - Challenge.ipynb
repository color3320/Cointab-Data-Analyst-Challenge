## Developing logic for assigning proper weight slab to each Order ID according to its Zone

```python
a = np.ceil(1.302/1.25)
b = a*1.25
b
```

## Implementing logic

```python
#Making function
#p = 'Delivery Zone as per X'
#q = 'Total weight as per X (KG)'
def wt_slab_cal(data,p,q):
    wt_slab = []
    no_of_slabs = []
    for value in range(len(data)):
        if data[f'{p}'][value] == 'a':
            a = np.ceil(data[f'{q}'][value]/0.25)
            x = a*0.25
            no_of_slabs.append(a)
            wt_slab.append(x)
        elif data[f'{p}'][value] == 'b':
            a = np.ceil(data[f'{q}'][value]/0.5)
            x = a*0.5
            no_of_slabs.append(a)
            wt_slab.append(x)
        elif data[f'{p}'][value] == 'c':
            a = np.ceil(data[f'{q}'][value]/0.75)
            x = a*0.75
            no_of_slabs.append(a)
            wt_slab.append(x)
        elif data[f'{p}'][value] == 'd':
            a = np.ceil(data[f'{q}'][value]/1.25)
            x = a*1.25
            no_of_slabs.append(a)
            wt_slab.append(x)
        elif data[f'{p}'][value] == 'e':
            a = np.ceil(data[f'{q}'][value]/1.5)
            x = a*1.5
            no_of_slabs.append(a)
            wt_slab.append(x)
        else:
            wt_slab.append('Invalid')
    return wt_slab, no_of_slabs
```

### Weight slab for X

```python
df_final['Weight slab as per X (KG)'], df_final['No of Slabs for X'] = wt_slab_cal(df_final, 'Delivery Zone as per X','Total weight as per X (KG)')
df_final
```

## Finding charge of each Order ID according to X

```python
estimate_charge_by_x = list()

No_of_slabs                           = df_final['No of Slabs for X']

Forward_Fixed_Charge                  = df_final['Forward Fixed Charge']
Forward_Additional_Weight_Slab_Charge = df_final['Forward Additional Weight Slab Charge']

RTO_Fixed_Charge                      = df_final['RTO Fixed Charge']
RTO_Additional_Weight_Slab_Charge     = df_final['RTO Additional Weight Slab Charge'] 

for value in range(len(df_final)):
    if df_final['Type of Shipment'][value] == 'Forward charges':
        
        result = ((No_of_slabs[value] - 1) * Forward_Additional_Weight_Slab_Charge[value]) + Forward_Fixed_Charge[value]        
        estimate_charge_by_x.append(round(result,1))  
        
    elif df_final['Type of Shipment'][value] == 'Forward and RTO charges':
        
        result = (((No_of_slabs[value] - 1) * Forward_Additional_Weight_Slab_Charge[value]) + Forward_Fixed_Charge[value]) 
        + (((No_of_slabs[value] - 1) * RTO_Additional_Weight_Slab_Charge[value]) + RTO_Fixed_Charge[value])        
        estimate_charge_by_x.append(round(result,1))
        
df_final['Expected Charge as per X (Rs.)'] = estimate_charge_by_x
df_final
```

## Difference between X's charge and Courier Company's Charge

```python
df_final['Difference Between Expected Charges and Billed Charges (Rs.)'] = df_final['Expected Charge as per X (Rs.)'] - df_final['Charges Billed by Courier Company (Rs.)']
df_final
```

## Creating Order Summary

```python
df = df_final[['Difference Between Expected Charges and Billed Charges (Rs.)','Expected Charge as per X (Rs.)']]
df
```

```python
amt = 0
ve_amt = 0
neg_amt = 0

count = 0
ve_count = 0
neg_count = 0

for value in range(len(df)):
    if df['Difference Between Expected Charges and Billed Charges (Rs.)'][value] == 0:
        amt += df['Expected Charge as per X (Rs.)'][value]
        count += 1
    elif df['Difference Between Expected Charges and Billed Charges (Rs.)'][value] < 0:
        neg_amt += df['Difference Between Expected Charges and Billed Charges (Rs.)'][value]
        neg_count += 1
    elif df['Difference Between Expected Charges and Billed Charges (Rs.)'][value] > 0:
        ve_amt += df['Difference Between Expected Charges and Billed Charges (Rs.)'][value]
        ve_count +=1  
```

```python
summary = pd.DataFrame({
    None :['Total Orders - Correctly Charged','Total Orders - Over Charged','Total Orders - Under Charged'],
   'Count':[count,neg_count,ve_count],
    'Amount':[amt,neg_amt,ve_amt]    
},columns=[None,'Count','Amount'])
summary
```
